```python
#use delimiter to split string ; in column countryname maxsplit=1
df['countryname'] = df['countryname'].str.split(';', n=1).str[0]
df.head()
```

```python

# keep only id	regionname	countryname	prodline	lendinginstr	lendinginstrtype	envassesmentcategorycode	supplementprojectflg	productlinetype	projectstatusdisplay	status	project_name	boardapprovaldate	board_approval_month	closingdate	lendprojectcost	ibrdcommamt	idacommamt	totalamt	grantamt	borrower	impagency	url
df = df[['id', 'regionname', 'countryname', 'prodline', 'lendinginstr', 'lendinginstrtype', 'envassesmentcategorycode', 'supplementprojectflg', 'productlinetype', 'projectstatusdisplay', 'status', 'project_name', 'boardapprovaldate', 'board_approval_month', 'closingdate', 'lendprojectcost', 'ibrdcommamt', 'idacommamt', 'totalamt', 'grantamt', 'borrower', 'impagency', 'url']]
df.head()
```

```python
#convert boardapprovaldate and closingdate to datetime
df['boardapprovaldate'] = pd.to_datetime(df['boardapprovaldate'])
df['closingdate'] = pd.to_datetime(df['closingdate'])
#check the data type
df.dtypes
```

```python
df['lendprojectcost'] = df['lendprojectcost'].str.replace(',', '').astype(float)
df['ibrdcommamt'] = df['ibrdcommamt'].str.replace(',', '').astype(float)
df['idacommamt'] = df['idacommamt'].str.replace(',', '').astype(float)
df['totalamt'] = df['totalamt'].str.replace(',', '').astype(float)
df['grantamt'] = df['grantamt'].str.replace(',', '').astype(float)
```

```python
#check data type
df.dtypes
```

## resolve missing values

```python
#check missing values
df.isnull().sum()
```

```python
df.shape()
```

```python
df.head()
```

```python
#remove missing valuies in supplementprojectflg, projectstatusdisplay, status
df = df.dropna(subset=['supplementprojectflg'])
df = df.dropna(subset=['projectstatusdisplay'])
df = df.dropna(subset=['status'])
```

```python
#fill missing values with mode
df['lendinginstr'] = df['lendinginstr'].fillna(df['lendinginstr'].mode()[0])
df['lendinginstrtype'] = df['lendinginstrtype'].fillna(df['lendinginstrtype'].mode()[0])
df['envassesmentcategorycode'] = df['envassesmentcategorycode'].fillna(df['envassesmentcategorycode'].mode()[0])
```

```python
#fix borrower, impagency column missing values with mode
df['borrower'] = df['borrower'].fillna(df['borrower'].mode()[0])
df['impagency'] = df['impagency'].fillna(df['impagency'].mode()[0])
```

```python
#handle missing values in lendprojectcost, ibrdcommamt, idacommamt, totalamt, grantamt
df['lendprojectcost'] = df['lendprojectcost'].fillna(df['lendprojectcost'].mean())
df['ibrdcommamt'] = df['ibrdcommamt'].fillna(df['ibrdcommamt'].mean())
df['idacommamt'] = df['idacommamt'].fillna(df['idacommamt'].mean())
df['totalamt'] = df['totalamt'].fillna(df['totalamt'].mean())
```

```python
#fill missing values in boardapprovaldate, board_approval_month, closingdate
df['boardapprovaldate'] = df['boardapprovaldate'].fillna(df['boardapprovaldate'].mode()[0])
df['board_approval_month'] = df['board_approval_month'].fillna(df['board_approval_month'].mode()[0])
df['closingdate'] = df['closingdate'].fillna(df['closingdate'].mode()[0])
```

```python
#fill missing values in datetime columns
df['boardapprovaldate'] = df['boardapprovaldate'].fillna(df['boardapprovaldate'].mode()[0])
df['closingdate'] = df['closingdate'].fillna(df['closingdate'].mode()[0])
df['board_approval_month'] = df['board_approval_month'].fillna(df['board_approval_month'].mode()[0])
```
